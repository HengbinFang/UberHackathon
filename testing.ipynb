```python
option = 1
rev = []

for geometry, transport in zip(
    g[g['option'] == option]["geometry"],
    g[g['option'] == option]["transport_mode"]):
    tmp = []
    
    for cd in geometry.coords:
        tmp.append([cd[1], cd[0]])
    rev.append(
        coordinates(
            tmp,
            transport
        )
    )

```
